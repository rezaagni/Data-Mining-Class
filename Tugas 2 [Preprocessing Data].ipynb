## Data Cleaning

```python
import pandas as pd
import numpy as np
```

```python
#import data
data = pd.read_csv('Echocardiogram.csv')
data
```

```python
data.head(15)
```

```python
data.tail(15)
```

```python
data.info()
```

```python
#statistika deskriptif
data.describe()
```

```python
np.sum(data.isnull())
```

```python
# imputasi data kontinu
data['Survival'] = data['Survival'].fillna((data['Survival'].mean())) 
data['Age-at-heart-attack'] = data['Age-at-heart-attack'].fillna((data['Age-at-heart-attack'].mean())) 
data['epss'] = data['epss'].fillna((data['epss'].mean())) 
data['lvdd'] = data['lvdd'].fillna((data['lvdd'].mean()))
data['Wall-motion-score'] = data['Wall-motion-score'].fillna((data['Wall-motion-score'].mean))
data['Wall-motion-index'] = data['Wall-motion-index'].fillna((data['Wall-motion-index'].mean))
data['Mult'] = data['Mult'].fillna((data['Mult'].mean))
data['Fractional-shortening'] = data['Fractional-shortening'].fillna((data['Fractional-shortening'].mean))


data.isnull().any()
```

```python
data.head()
```

```python
#imputasi data kategorik
data['Still-alive'] = data['Still-alive'].fillna((data['Still-alive'].value_counts().index[0]))
data['Pericardial-effusion'] = data['Pericardial-effusion'].fillna((data['Pericardial-effusion'].value_counts().index[0]))
data['group'] = data['group'].fillna((data['group'].mode))
data['Alive-at-1'] = data['Alive-at-1'].fillna((data['Alive-at-1'].value_counts().index[0]))
data.isnull().any()
```

```python
data.head()
```

## Encode Labels

```python
from sklearn.preprocessing import LabelEncoder
```

```python
# Encode untuk Still-alive, Pericardial-effusion, Alive-at-1 tanpa membuat kolom baru
data['Still-alive'] = LabelEncoder().fit_transform(data['Still-alive'])
data['Pericardial-effusion'] = LabelEncoder().fit_transform(data['Pericardial-effusion'])
data['Alive-at-1'] = LabelEncoder().fit_transform(data['Alive-at-1'])
data.head()
```

## Standardize

```python
data.info()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
data['Survival'] = StandardScaler().fit(data[['Survival']]).transform(data[['Survival']])
data['Age-at-heart-attack'] = StandardScaler().fit(data[['Age-at-heart-attack']]).transform(data[['Age-at-heart-attack']])
data['epss'] = StandardScaler().fit(data[['epss']]).transform(data[['epss']])
data['lvdd'] = StandardScaler().fit(data[['lvdd']]).transform(data[['lvdd']])

data.head()
```

## Normalize

```python
from sklearn.preprocessing import Normalizer 
```

```python
data['Survival'] = Normalizer ().fit(data[['Survival']]).transform(data[['Survival']])
data['Age-at-heart-attack'] = Normalizer ().fit(data[['Age-at-heart-attack']]).transform(data[['Age-at-heart-attack']])
data['epss'] = Normalizer ().fit(data[['epss']]).transform(data[['epss']])
data['lvdd'] = Normalizer ().fit(data[['lvdd']]).transform(data[['lvdd']])
data.head()
```
